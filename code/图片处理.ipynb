```python
import os
import shutil
import functools
import pandas as pd
import numpy as np
def sort_fig_name(name1, name2):
    fig_name1 = int(name1[:-4])
    fig_name2 = int(name2[:-4])
    if fig_name1 > fig_name2:
        return 1
    if fig_name1 < fig_name2:
        return -1
    else:
        return 0

def pix_to_mm(data):
    """将头部喷流数据像素转为毫米"""
    _ = data["头锥直径像素"] / data["头锥直径"]
    
    return round(data["最大脱体距离像素"] / _, 1)

def hengxiang_pix_to_mm(data):
    """将横向喷流数据像素转化为毫米"""
    _ = data["斜劈高像素"] / data["斜劈高"]
    return round(data["激波高度像素"] / _, 1)

def get_group_hengxiang(data, group_origin_idx):
    """从data中找出想要的组"""
    index = [True if i in range(group_origin_idx,group_origin_idx+45) else False for i in range(0, data.values.shape[0])]
    return index
    
    
```

# 处理头部喷流序列数据

```python
rootpath = "./1-2 头部喷流--控制过程图像序列/"
with open(rootpath + "data.csv") as f:
    data = pd.read_csv(f)

data["最大脱体距离"] = pix_to_mm(data) # 将像素转化为毫米
columns = data.columns
remove_cols = ["电极间距", "击穿电压", "头锥直径像素", "最大脱体距离像素"]
retain_cols = [col for col in columns if col not in remove_cols]
data = data[retain_cols]
data.to_csv(rootpath + "final_data.csv", index = False)
```

```python
with open("./origin_data_hengxiang.csv") as f:
    data = pd.read_csv(f)
data["激波高度"] = hengxiang_pix_to_mm(data)
columns = data.columns
remove_cols = ["斜劈高像素", "激波高度像素"]
retain_cols = [col for col in columns if col not in remove_cols]
data = data[retain_cols]
data.to_csv("./final_data_hengxiang.csv", index = False)
```

# 分组划分

```python
group_start_idx = range(0, 1260, 45) # 每一组的开始索引
count = 1
for i in group_start_idx:
    group_index = get_group_hengxiang(data, i)
    not_group_index = [not id for id in group_index]
    group_data = data.iloc[group_index, :]
    group_data_rest = data.iloc[not_group_index, :]
    savepath1 = "./hengxiang_group" + str(count) + ".csv"
    savepath2 = "./hengxiang_group" + str(count) + "_rest.csv"
    group_data.to_csv(savepath1, index = False)
    group_data_rest.to_csv(savepath2, index = False)
    count += 1
```
